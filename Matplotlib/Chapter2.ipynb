# 第2章 使用统计函数绘制简单图形

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## 2.1 函数bar() - 用于绘制柱状图

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]

plt.bar(x,y,align="center",color="c",tick_label=["q","a","c","e","r","j",",b","p"],
        hatch="/")

plt.xlabel("Box Name")
plt.ylabel("Box Weight(kg)")

plt.show()
```

## 2.2 函数barh() - 用于绘制条形图

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]

plt.barh(x,y,align="center",color="c",tick_label=["q","a","c","e","r","j",",b","p"],
        hatch="/")

plt.ylabel("Box Name")
plt.xlabel("Box Weight(kg)")

plt.show()
```

## 2.3 函数hist() - 用于绘制直方图

```python
boxWeight = np.random.randint(0,10,100)

x = boxWeight

bins = range(0,11,1)

plt.hist(x,bins=bins,
         color="g",
         histtype="bar",
         rwidth=1,
         alpha=0.6,
         ec='black')

plt.xlabel("Box Weight(kg)")
plt.ylabel("Sell Number")

plt.xlim(0,10)

plt.show()
```

## 2.4 函数pie() - 用于绘制饼图

```python
kinds = ["A","B","C","D"]

colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

soldNums = [0.05,0.45,0.15,0.35]

plt.pie(soldNums,
        labels=kinds,
        autopct="%3.1f%%",
        startangle=60,
        colors=colors)

plt.title("Different Boxes Sell Situation")

plt.show()
```

## 2.5 函数polar() - 用于绘制极线图

```python
barSlices = 12

theta = np.linspace(0,2*np.pi,barSlices,endpoint=False)
r = 30*np.random.rand(barSlices)

plt.polar(theta,r,
          color="chartreuse",
          linewidth=2,
          marker='*',
          mfc='b',
          ms=10)

plt.show()
```

## 2.6 函数scatter() - 用于绘制气泡图

```python
a = np.random.randn(100)
b = np.random.randn(100)

plt.scatter(a,b,s=np.power(10*a+20*b,2),
            c = np.random.rand(100),
            cmap=mpl.cm.RdYlBu,
            marker='o')

plt.show()
```

## 2.7 函数stem() - 用于绘制棉棒图

```python
x = np.linspace(0.5,2*np.pi,20)
y = np.random.randn(20)

plt.stem(x,y,linefmt='-.',markerfmt='o',basefmt='-')

plt.show()
```

## 2.8 函数boxplot() - 用于绘制箱线图

```python
x = np.random.randn(1000)

plt.boxplot(x)

plt.xticks([1],["Random Number Creator AlphaRM"])
plt.ylabel("Random Number")
plt.title("Stability of Random Number Creator")

plt.grid(axis="y",ls=":",lw=1,c='g',alpha=0.4)

plt.show()
```

## 2.9 函数errorbar() - 用于绘制误差棒图

```python
x = np.linspace(0.1,0.6,6)
y = np.exp(x)

plt.errorbar(x,y,fmt="bo:",yerr=0.2,xerr=0.02)

plt.xlim(0,0.7)

plt.show()
```
