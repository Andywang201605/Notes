# 第3章 绘制统计图形

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

## 3.1 柱状图
柱状图主要应用在定性数据的可视化场景中，或者离散型数据的分布展示

```python
# some simple data
x = [1,2,3,4,5]
y = [6,10,4,5,1]

# create bar
plt.bar(x,y,align="center",color="b",tick_label=["A","B","C","D","E"],
        alpha=0.6)

# set x,y_axis label
plt.xlabel("Difficulty")
plt.ylabel("Number")

# set yaxis grid
plt.grid(True,axis='y',ls=':',color='r',alpha=0.3)

plt.show()
```

## 3.2 条形图 （水平柱状图）

```python
# some simple data
x = [1,2,3,4,5]
y = [6,10,4,5,1]

# create bar
plt.barh(x,y,align="center",color="c",tick_label=["A","B","C","D","E"],
        alpha=0.6)

# set x,y_axis label
plt.ylabel("Difficulty")
plt.xlabel("Number")

# set yaxis grid
plt.grid(True,axis='x',ls=':',color='r',alpha=0.3)

plt.show()
```

## 3.3 堆积图 （多个柱状图）
### 3.3.1 堆积柱状图

```python
# some simple data
x = [1,2,3,4,5]
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

# create bar
plt.bar(x,y,align="center",color="#66c2a5",tick_label=["A","B","C","D","E"],
        label='Class A')
plt.bar(x,y1,align="center",bottom=y,color="#8da0cb",label="Class B")

# set x,y_axis label
plt.xlabel("Difficulty")
plt.ylabel("Number")

# set yaxis grid
plt.grid(True,axis='y',ls=':',color='r',alpha=0.3)

plt.legend()

plt.show()
```

### 3.3.2 堆积条形图

```python
# some simple data
x = [1,2,3,4,5]
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

# create bar
plt.barh(x,y,align="center",color="#66c2a5",tick_label=["A","B","C","D","E"],
        label="Class A")
plt.barh(x,y1,align="center",left=y,color="#8da0cb",label="Class B")

# set x,y_axis label
plt.ylabel("Difficulty")
plt.xlabel("Number")

# set yaxis grid
plt.grid(True,axis='x',ls=':',color='r',alpha=0.3)

plt.legend()

plt.show()
```

## 3.4 分块图
将多组数据以堆积图的形式进行展示

### 3.4.1 多数据并列柱状图

```python
# some simple data
x = np.arange(5)
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

bar_width = 0.35
tick_label = ["A","B","C","D","E"]

# create bar
plt.bar(x,y,bar_width,color="c",align="center",label="Class A",alpha=0.5)
plt.bar(x+bar_width,y1,bar_width,color="b",align="center",label="Class B",
        alpha=0.5)

# set x,y_axis label
plt.xlabel("Difficulty")
plt.ylabel("Number")

# set xaxis ticks and ticklabels
plt.xticks(x+bar_width/2,tick_label)

plt.legend()

plt.show()
```

### 3.4.2 多数据平行条形图

```python
# some simple data
x = np.arange(5)
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

bar_width = 0.35
tick_label = ["A","B","C","D","E"]

# create horizontal bar
plt.barh(x,y,bar_width,color="c",align="center",label="Class A",alpha=0.5)
plt.barh(x+bar_width,y1,bar_width,color="b",align="center",label="Class B",
        alpha=0.5)

# set x,y_axis label
plt.ylabel("Difficulty")
plt.xlabel("Number")

# set xaxis ticks and ticklabels
plt.yticks(x+bar_width/2,tick_label)

plt.legend()

plt.show()
```

## 3.5 柱状图装饰

```python
# some simple data
x = [1,2,3,4,5]
y = [6,10,4,5,1]

# create bar
plt.bar(x,y,align="center",color="b",tick_label=["A","B","C","D","E"],
        hatch="///")

# hatch can be chosen from "/" "\\" "|" "-" etc.

# set x,y_axis label
plt.xlabel("Difficulty")
plt.ylabel("Number")

# set yaxis grid
plt.grid(True,axis='y',ls=':',color='r',alpha=0.3)

plt.show()
```

## 3.6 堆积折线图、间断条形图和阶梯图

### 3.6.1 用函数stackplot()绘制堆积折线图
堆积折线图是通过绘制不同数据集的折线图而生成的

```python
x = np.arange(1,6,1)
y = [0,4,3,5,6]
y1 = [1,3,2,4,7]
y2 = [3,4,1,6,5]

labels = ["BluePlanet","BrownPlanet","GreenPlanet"]
colors = ["#8da0cb","#fc8d62","#66c2a5"]
          
plt.stackplot(x,y,y1,y2,labels=labels,colors=colors)

plt.xlim(1,5)
plt.legend(loc="upper left")

plt.show()
```

### 3.6.2 用函数broker_barh()绘制间断条形图
主要用来可视化定性数据的相同指标在时间维度上的指标值的变化情况（在画布上绘制从某一点绘制定长定宽的矩形区域）

```python
plt.broken_barh([(30,100),(180,50),(260,70)],(20,8),facecolors="#1f78b4")
plt.broken_barh([(60,90),(190,20),(230,30),(280,60)],(10,8),facecolors=('#7fc97f',
                 '#beaed4','#fdc086','#ffff99'))

plt.xlim(0,360)
plt.ylim(5,35)
plt.xlabel('Time')

plt.xticks(np.arange(0,361,60))
plt.yticks([15,25],["Class A","Class B"])

plt.grid(ls='-',lw=1,color="gray")

plt.title("Time allocation for different class")

plt.show()
```

### 3.6.3 用函数step()绘制阶梯图
通常用于实践序列数据的可视化任务中，凸显时序数据的波动周期和规律

```python
x = np.linspace(1,10,10)
y = np.sin(x)

plt.step(x,y,color="#8dd3cf",where='pre',lw=2)

# where 可取值有 'pre' 'post'，代表数据相邻数据点是左开右闭还是左闭右开

plt.xlim(0,11)
plt.xticks(np.arange(1,11,1))
plt.ylim(-1.2,1.2)

plt.show()
```

## 3.7 直方图
直方图主要用于连续型数据的可视化展示

### 3.7.2 绘制原理

```python
# set test scores
scoresT = np.random.randint(0,100,100)

x = scoresT

# plot histogram
bins = range(0,101,10)

plt.hist(x,bins=bins,color='#377eb8',histtype='bar',
         rwidth=1.0,edgecolor='black')

# rwidth 是柱体的相对宽度，取值范围0-1
         
# set x,y-axis label
plt.xlabel('Grade')
plt.ylabel('Number')

plt.xlim(0,100)

plt.show()
```

### 3.7.4 堆积直方图

```python
# set test scores
scoresT1 = np.random.randint(0,100,100)
scoresT2 = np.random.randint(0,100,100)

x = [scoresT1,scoresT2]
colors = ['#8dd3cf','#bebada']
labels = ['Class A','Class B']

# plot histogram
bins = range(0,101,10)

plt.hist(x,bins=bins,color=colors,histtype='bar',
         rwidth=1.0,stacked=True,label=labels,edgecolor='black')

# plt.hist(x,bins=bins,color=colors,histtype='bar',
#          rwidth=1.0,stacked=False,label=labels,edgecolor='black')
# plt.hist(x,bins=bins,color=colors,histtype='stepfilled',
#          rwidth=1.0,stacked=True,label=labels,edgecolor='black')

# stacked = True即绘制堆积直方图，如果为False则为并列堆叠，Ctrl+/注释上述语句可以尝试运行
# histtype 为直方图样式，主要有bar和stepfilled的两种样式

# set x,y-axis label
plt.xlabel('Grade')
plt.ylabel('Number')

plt.xlim(0,100)

plt.title('Grade Distribution in Different Classes')

plt.legend(loc='upper left')

plt.show()
```

## 3.8 饼图
用于对数据进行比例展示

```python
labels = ['A','B','C','D']

students = [0.35,0.15,0.20,0.30]

colors = ['#377eb8','#4daf4a','#984ea3','#ff7f00']
          
explode = (0.1,0.1,0.1,0.1)

# exploded pie chart
plt.pie(students,explode=explode,labels=labels,autopct='%3.1f%%',
        startangle=45,shadow=True,colors=colors)

# 为了设置百分比和标签的位置可以利用参数pctdistance和labeldistance设置，其值都是相对半径的比例

# plt.pie(students,labels=labels,autopct='%3.1f%%',
#         startangle=45,shadow=True,colors=colors,pctdistance=0.7,labeldistance=1.2)



plt.title('Product Distribution')

plt.show()
```

### 3.8.4 应用 -- 内嵌环形饼图

```python
elements = ['Flour','Sugar','Cream','Jam','Nut']

weight1 = [40,15,20,10,15]
weight2 = [30,25,15,20,10]

colormapList = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00']
outer_colors = colormapList
inner_colors = colormapList

wedges1,texts1,autotexts1 = plt.pie(weight1,autopct='%3.1f%%',
                                    radius=1,pctdistance=0.85,
                                    colors=outer_colors,
                                    textprops=dict(color='w'),
                                    wedgeprops=dict(width=0.6,edgecolor='w'))

wedges2,texts2,autotexts2 = plt.pie(weight2,autopct='%3.1f%%',
                                    radius=0.7,pctdistance=0.75,
                                    colors=inner_colors,
                                    textprops=dict(color='w'),
                                    wedgeprops=dict(width=0.3,edgecolor='w'))

# plt.pie的返回值有三个，其中wedge是饼图中的每一片小饼，text是label文本，autotext为百分比文本
# textprops为一个字典，为图中文本的格式；wedgeprops为图中饼的格式，width为饼的宽度，比如radius=1，而wedge的width为0.3
# 那么饼（环）图就是从0.7-1.0的一个环

plt.legend(wedges1,elements,fontsize=12,
           title='Ingredient',loc='center left',
           bbox_to_anchor=(0.91,0,0.3,1))

plt.setp(autotexts1,size=8,weight="bold")
plt.setp(autotexts2,size=8,weight="bold")

plt.title("Different Bread Ingredients")

plt.show()
```

## 3.9 箱线图
用于展示一组（大量）数据的分布情况

### 3.9.2 绘制原理

```python
testA = np.random.randn(5000)
testB = np.random.randn(5000)

testList = [testA,testB]
labels = ['Mechine Alpha','Mechine Beta']
colors = ['#1b9e77','#d95f02']

whis = 1.6; width = 0.35

bplot = plt.boxplot(testList,whis=whis,widths=width,
                    sym='o',labels=labels,patch_artist=True)

# whis用于确定箱须的位置，箱须位于Q1 - 1.5 * IQR和Q3 + 1.5 * IQR
# patch_artist用于确定箱体颜色，True为上色，如果为False则下面的语句会报错
# sym为离群值的样式，其取值与marker一致
# width为箱体的宽度

for patch,color in zip(bplot['boxes'],colors):
    patch.set_facecolor(color)
    
plt.ylabel('Product Quality')
plt.title('Product Made by Different Machine')

plt.grid(axis='y',ls=':',lw=1,color='gray',alpha=0.4)

plt.show()
```

### 3.9.4 箱线图应用 -- 水平方向的箱线图

```python
x = np.random.randn(1000)

plt.boxplot(x,vert=False)

# vert = False为设置箱线图为横向
# 如果只关注大部分数据的分布，可以不绘制离群值，用showfliers参数进行控制，默认为True即绘制离群点

# plt.boxplot(x,vert=False,showfliers=False)

plt.xlabel('Product Quality')
plt.yticks([1],['Machine Alpha'],rotation=90)
plt.title('Machine Performance')

plt.grid(axis='x',ls=':',lw=1,color='gray',alpha=0.4)

plt.show()
```

## 3.10 误差棒图

### 3.10.2 绘制原理

```python
x = np.linspace(0.1,0.6,10)
y = np.exp(x)

error = 0.05 + 0.15 * x

lower_error = error
upper_error = 0.3*error
error_limit = [lower_error,upper_error]

plt.errorbar(x,y,yerr=error_limit,fmt='o:', # fmt与plot(x,y)中的线形和marker的样式一致
             ecolor='y',elinewidth=4,       # 本行为errorline的样式
             ms=5,mfc='c',mec='r',          # 本行为markersize,markerfacecolor,markeredgecolor
             capthick=1,capsize=2)          # 本行为误差棒边界横杠的厚度和大小

plt.xlim(0,0.7)

plt.show()
```

### 3.10.3 带误差棒的柱状图

```python
x = np.arange(5)
y = [100,68,79,91,82]
std_err = [7,2,6,10,5]

error_attri = dict(elinewidth=2,ecolor='black',capsize=3)

plt.bar(x,y,color='c',width=0.6,align='center',
        yerr=std_err,error_kw=error_attri,
        tick_label=['A','B','C','D','E'])

# yerr用于控制误差，也可以做成不对称的（如上例）；error_kw用于控制误差棒的样式
# 如果为条形图（barh），设置xerr即可
# 也可以用柱状图和误差棒图联合绘制（为了不显示连线需要设置linestype(或ls)为'none'）

# plt.bar(x,y,color='c',width=0.6,align='center',
#         tick_label=['A','B','C','D','E'])
# plt.errorbar(x,y,yerr=std_err,ls='none',**error_attri)

plt.xlabel('Garden')
plt.ylabel('Tourists')

plt.title('Number of Tourists for Different Garden')

plt.grid(True,axis='y',ls=':',color='gray',alpha=0.2)

plt.show()
```

### 3.10.6 带误差的堆积柱状图

```python
x = np.arange(5)
y1 = [1200,2400,1800,2200,1600]
y2 = [1050,2100,1300,1600,1340]
std_err1 = [150,100,180,130,80]
std_err2 = [120,110,170,150,120]

bar_width = 0.6
tick_label = ['Jan','Feb','Mar','Apr','May']
error_attri = dict(ecolor='black',elinewidth=2,capsize=0)

plt.bar(x,y1,bar_width,color='#6495ED',align='center',
        yerr=std_err1,error_kw=error_attri,
        label='City 1')

plt.bar(x,y2,bar_width,color='#FFA500',align='center',
        bottom = y1,yerr=std_err1,error_kw=error_attri,
        label='City 2')

# 带误差的并列柱状图同理，需要用x与width来控制即可

plt.xlabel('Month')
plt.ylabel('Tourists')

plt.title('Tourists in Different City in Each Month')

plt.grid(True,axis='y',ls=':',color='gray',alpha=0.2)

plt.xticks(x,tick_label)

plt.legend()

plt.show()
```
