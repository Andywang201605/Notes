# 第4章 完善统计图形

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

## 4.1 添加图例（legend）和标题（title）

### 4.1.1 设置方法

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
y = np.sin(x)
y1 = np.cos(x)

plt.plot(x,y,label=r'$\sin(x)$')
plt.plot(x,y1,label=r'$\cos(x)$')

# r'$$'使得Python可以用TeX规范对文本内容进行解析

plt.legend(loc='lower left')

plt.title('Sinine and Cosine Function')

plt.show()
```

### 4.1.2 图例的展示样式的调整

```python
x = np.arange(0,2.1,0.1)
y = np.power(x,3)
y1 = np.power(x,2)
y2 = np.power(x,1)

plt.plot(x,y,ls='-',lw=2,label='$x^3$')
plt.plot(x,y1,ls='-',lw=2,c='r',label='$x^2$')
plt.plot(x,y2,ls='-',lw=2,c='y',label='$x^1$')

plt.legend(loc='upper left',bbox_to_anchor=(0.05,0.95),ncol=3,
           title='power function',shadow=True,fancybox=True)

# loc为图例所在的位置一般为'upper','center','lower'和'left','right'的组合，同样还有'best'参数
# bbox_to_anchor为线框位置参数，以Axes为相对值为（线框左侧距离左侧的相对x轴的距离，线框下侧距离下侧相对y轴的距离，
# 线框横向长度相对x轴的长度，线框纵向长度相对y轴的长度）
# ncol为图例的列数，默认为1
# shadow用于控制阴影，fancybox用于控制图例矩形的角

plt.show()
```

### 4.1.3 标题的展示样式的调整

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)

plt.plot(x,y,ls='-',lw=2,color='g')

plt.title('center demo')
plt.title('left demo',loc='left',fontdict=dict(size='xx-large',color='r',
                                               family='Times New Roman'))
plt.title('right demo',loc='right',family='Comic Sans MS',
          size=20,style='oblique',color='c')

plt.show()
```

### 4.1.4 带图例的饼图

```python
elements = ['Flavor','Sugar','Cream','Jam','Nut']

weight = [40,15,20,10,15]

colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00']

wedges, texts, autotexts = plt.pie(weight, autopct='%3.1f%%',
                                   colors=colors)

plt.legend(wedges, elements, fontsize=12,
           title='Ingredient', loc='center left',
           bbox_to_anchor=(0.91,0,0.3,1))

plt.setp(autotexts,size=12,color='w',weight='bold')

plt.show()
```

## 4.2 调整刻度范围和刻度标签

### 4.2.1 调整刻度范围和刻度标签的方法

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
y = np.sin(x)

plt.subplot(211)

plt.plot(x,y)

plt.subplot(212)

plt.xlim(-2*np.pi,2*np.pi)

plt.xticks([-2*np.pi,-3*np.pi/2,-1*np.pi,-1*np.pi/2,0,np.pi/2,np.pi,
            3*np.pi/2,2*np.pi],

[r'$-2\pi$',r'$-3\pi/2$',r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi$',
 r'$3\pi/2$',r'$2\pi$'])

plt.plot(x,y)

plt.show()
```

### 4.2.3 逆序设置坐标轴刻度标签

```python
time = np.arange(1,11,0.5)
Power = np.power(time,2)+0.7

plt.plot(time,Power,linestyle='-',linewidth=2,color='r')

plt.xlim(10,1)

plt.xlabel('Year')
plt.ylabel('Power')

plt.title('Machine Power')

plt.grid(ls=':',lw=1,color='gray',alpha=0.5)

plt.show()
```

## 4.3 向统计图形添加表格

```python
labels = ['A','B','C','D']

students = [0.35,0.15,0.20,0.30]
explode = (0.1,0.1,0.1,0.1)
colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

plt.pie(students, explode=explode, labels=labels,
        autopct='%1.1f%%',startangle=45,shadow=True,
        colors=colors)

plt.title('Test Difficulty Distribution')

# Table
colLables = ['A','B','C','D']
rowLabels = ['Number']
values = [[350,150,200,300]]
colColors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']
             
plt.table(cellText=values, cellLoc='center', colWidths=[0.1]*4,
          colLabels=colLables,rowLabels=rowLabels,colColours=colColors,
          rowLoc='center',loc='bottom')

# cellText为二维数组，代表每一个单元格中的内容； cellLoc为单元格中数值的对齐位置
# colWidths是表格每列的宽度； rowLoc为行名称的对齐方式； loc为表格在画布中的位置

plt.show()
```
